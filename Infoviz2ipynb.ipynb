```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 500)
```

```python
df = pd.read_excel("/content/GuttmacherInstituteAbortionDataByState.xlsx", header = 0)
```

```python
df.isnull().sum().sort_values(ascending = False)
```

Which states in the U.S have the greatest barriers to abortion acceess, as measured by the percentage of counties without a known abortion clinic in 2020 and the percentage of women aged 15-44 livin in those countries?

```python
earnest_df = df[['U.S. State', '% of counties without a known clinic, 2020', '% of women aged 15-44 living in a county without a clinic, 2020']]
```

```python
earnest_df = earnest_df.sort_values('% of women aged 15-44 living in a county without a clinic, 2020', ascending=False).head(10)
```

```python
earnest_df.head()
```

```python
earnest_df.isnull().sum()
```

* https://stackoverflow.com/questions/23688227 confusion-about-artist-place-in-matplotlib-hierarchy
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html
* https://www.geeksforgeeks.org/python/adding-value-labels-on-a-matplotlib-bar-chart/
* https://stackoverflow.com/questions/42281851/how-to-add-padding-to-a-plot-in-python


```python
fig, ax = plt.subplots(figsize = (14, 8))

sns.barplot(
    data = earnest_df.head(10),
    y = "U.S. State",
    x = "% of women aged 15-44 living in a county without a clinic, 2020",
    palette = "RdYlBu_r",
    ax = ax
)

for container in ax.containers:
  ax.bar_label(container, fmt='%.0f%%', padding = 3)


ax.set_xlabel("% of women aged 15-44 living in a state without a clinic, 2020",
              fontsize = 12, fontweight="bold")
ax.set_ylabel("State", fontsize = 12, fontweight="bold")
ax.set_title('Top 10 States: Percentage of Women Living in States Without Abortion Clinics (2020)\nData Source: Guttmacher Institute',
             fontsize=14, fontweight='bold', pad=20)


ax.set_xlim(0, 100)

ax.grid(axis = "x", alpha = 0.3, linestyle = "--")

plt.tight_layout()
plt.show()
```
